# Creating data

```python
movie = ['JL','Batman','Avengers','Ironman','Fast&Furious']
percentage = [3,8,9,7,9]
```

# importing Libaraies

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# scatter plot

```python
plt.scatter(movie,percentage)
```

# line plot

```python
plt.plot(movie, percentage,linestyle='solid')
#dashed, dashdot,dotted
```

```python
plt.pie(percentage,labels=movie)
```

# Bar chart

```python
plt.bar(movie,percentage)
```

```python
plt.bar(percentage,movie)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sc
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering
```

```python
#import iris data
iris =  datasets.load_iris()
```

```python
iris
```

```python
iris_data = pd.DataFrame(iris.data )
```

```python
iris_data 
```

```python
iris_data.columns = iris.feature_names
```

```python
iris_data["Flower_type"] = iris.target
```

```python
iris_data
```

```python
iris_X = iris_data.iloc[:,[0 ,1 ,2 ,3]].values
```

```python
iris_Y = iris_data.iloc[:, 4].values
```

```python
plt.figure(figsize=(10,7))
plt.scatter(iris_X[iris_Y == 0,0],iris_X[iris_Y == 0,1], s = 100,c='blue',label='set-1')
plt.scatter(iris_X[iris_Y == 1,0],iris_X[iris_Y == 1,1], s = 100,c='red',label='set-2')
plt.scatter(iris_X[iris_Y == 2,0],iris_X[iris_Y == 2,1], s = 100,c='green',label='set-3')
plt.legend()
plt.xlabel('Sample Index')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
# plot dendrogram
plt.figure(figsize = (20,7))
plt.title("Dendrograms")
```

```python
sc.dendrogram(sc.linkage(iris_X,method = 'ward'))
```

```python
plt.title("Dendrograms")
plt.xlabel("Sample Index")
plt.ylabel("Euclidean Distance")
```
